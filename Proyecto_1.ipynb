```python
import folium
# Impresión de la versión de Folium
print(folium.__version__)

import pandas as pd

print(pd.__version__)


```

```python
# CREACIÓN DE MAPA CON TESELAS

# Creación de un mapa
map= folium.Map(
    location=[10,-84],
    width = 650, heigth = 400,
    zoom_start = 7,
    tiles = 'CartoDB dark_matter',
    control_scale=True)

folium.TileLayer(
   tiles='CartoDB positron', 
   name='CartoDB positron').add_to(map)

folium.TileLayer(
    tiles='Stamen Terrain', 
    name='Stamen Terrain').add_to(map)


folium.LayerControl().add_to(map)

map

```

```python
# Biblioteca para funciones del sistema operativo

import os

# Rutas a los archivos GeoJson
cantones_geo = os.path.join('datos/', 'cr_limite_cantonal_ign_wgs84.geojson')
distritos_geo = os.path.join('datos/', 'cr_limite_distrital_ign_wgs84.geojson')


# Creación del mapa
map = folium.Map(location=[10, -84], tiles='CartoDB dark_matter', 
                 zoom_start=7,control_scale=True,)


folium.TileLayer(
   tiles='CartoDB positron', 
   name='CartoDB positron').add_to(map)

folium.TileLayer(
    tiles='Stamen Terrain', 
    name='Stamen Terrain').add_to(map)

folium.GeoJson(data=cantones_geo, name='Cantones').add_to(map)

# Control de capas

folium.LayerControl().add_to(map)

# Despliegue del mapa

map


```
